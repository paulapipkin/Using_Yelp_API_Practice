```python
import json
import pandas as pd
from yelpapi import YelpAPI
```

```python
#import access info from file
with open('/Users/paula/.secret/yelp_api.json') as f:
          login = json.load(f)
login.keys()
```

```python
#create instance(like an ID in this case) to login

yelp_access = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_access 
```

```python
query_1 = yelp_access.search_query(location='Ventura,CA', term='beer')

type(query_1)
```
